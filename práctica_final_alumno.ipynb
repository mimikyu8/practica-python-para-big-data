```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']


# A. Una función que calcule el precio total de la compra
def total_compra(articulos, compra):
    comprado = []
    list(map(lambda x: comprado.append(articulos[x]), compra))
    return reduce(lambda x, y: x + y, comprado)

print(total_compra(articulos, compra))

# B. Una función que calcule el precio total de la compra con un descuento del 10%
def descuento_10(articulos, compra):
    return total_compra(articulos, compra) * 0.9

print(descuento_10(articulos, compra))

# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€
def descuento_mayor_100(articulos, compra):
    if total_compra(articulos, compra) > 100:
        return descuento_10(articulos, compra)
    else:
        return total_compra(articulos, compra)

print(descuento_mayor_100(articulos, compra))

# D. Una función que calcule el precio total aplicando el IVA (21%)
def calcular_iva(articulos, compra):
    return total_compra(articulos, compra) * 0.21

print(calcular_iva(articulos, compra))
# E.lista los artículos cuyo precio es superior a 20€
def mayor_20(articulos):
    return dict(filter(lambda x: x[1] > 20, articulos.items()))

print(mayor_20(articulos))
```

```python
'''
14. Dada una lista de tuplas con el nombre de un alumno, apellidos, curso y sus notas. 

 Define una funcion que reciba el curso y saque una lista en la que aparezca e nombre y apellidos y el promedio de sus notas.
 Puedes definir más funciones si lo consideras necesario.
'''

alumnos = [('Juan', 'Pérez', '1', [5, 6, 7, 8, 9]),
            ('Ana', 'García', '2', [5, 6, 7, 8, 9]),
            ('Luis', 'González', '1', [5, 6, 7, 8, 9]),
            ('María', 'Martínez', '2', [5, 6, 7, 8, 9]),
            ('Pedro', 'Rodríguez', '1', [5, 6, 7, 8, 9]),
            ('Lucía', 'Hernández', '2', [5, 6, 7, 8, 9]),
            ('Marta', 'López', '1', [5, 6, 7, 8, 9]),
            ('Sara', 'Díaz', '2', [5, 6, 7, 8, 9]),
            ('Javier', 'Sánchez', '1', [5, 6, 7, 8, 9]),
            ('Miguel', 'Fernández', '2', [5, 6, 7, 8, 9]),
            ('Sergio', 'Jiménez', '1', [5, 6, 7, 8, 9]),
            ('Sandra', 'Ruiz', '2', [5, 6, 7, 8, 9]),
            ('Pablo', 'Álvarez', '1', [5, 6, 7, 8, 9]),
            ('María', 'Gómez', '2', [5, 6, 7, 8, 9]),
]

def info_alumnos(curso, alumnos):
    nombre = []
    apellido = []
    cursando = []
    notas = []

    i = 0
    while i < len(alumnos):
        nombre.append(alumnos[i][0])
        apellido.append(alumnos[i][1])
        cursando.append(alumnos[i][2])
        notas.append(alumnos[i][3])
        i += 1
    
    promedio = []
    for nota in notas:
        promedio.append(reduce(lambda x, y: x + y, nota, 0)/(len(notas)))

    info = []
    for j in range(len(alumnos)):
        if cursando[j] == curso:
            info.append([nombre[j], apellido[j], promedio[j]])
        else:
            continue
    return info
```
